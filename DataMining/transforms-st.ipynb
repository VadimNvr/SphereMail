```python
import pylab as pl
import numpy as np

%pylab inline
```

```python
X = np.array([np.random.lognormal(size=100), np.random.uniform(size=100)]).T
```

```python
pl.figure(figsize=(5,5))
pl.scatter(X[:, 0], X[:, 1], color="red")
pl.title("Original")
pl.show()
```

```python
X[:, 0] = (X[:, 0] - X[:, 0].min()) / (X[:, 0].max() - X[:, 0].min())                            
```

```python
pl.figure(figsize=(5,5))
pl.scatter(X[:, 0], X[:, 1], color="blue")
pl.title("Standartized")
pl.show()
```

```python
pl.figure(figsize=(5,5))
pl.scatter(numpy.log(X[:, 0]), X[:, 1], color="blue")
pl.title("Log")
pl.show()
```

```python
pl.figure(figsize=(5,5))
pl.scatter(np.log(X[:, 0] + 1), X[:, 1], color="blue")
pl.title("SQRT")
pl.show()
```
