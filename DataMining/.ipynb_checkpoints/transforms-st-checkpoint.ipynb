```python
import pylab as pl
import numpy as np

%pylab inline
```

```python
X = np.array([np.random.lognormal(size=100), np.random.uniform(size=100)]).T
```

```python
pl.figure(figsize=(5,5))
pl.scatter(X[:, 0], X[:, 1], color="red")
pl.title("Original")
pl.show()
```

```python
pl.figure(figsize=(5,5))
pl.scatter((X[:, 0] - X[:, 0].min()) / (X[:, 0].max() - X[:, 0].min()), X[:, 1], color="blue")
pl.title("Normalized")
pl.show()
```

```python
pl.figure(figsize=(5,5))
#вычесть среднее и поделить на корень от стандартного отклонения
pl.scatter((X[:, 0] - np.mean(X[:, 0])) / numpy.sqrt(numpy.std(X[:, 1])), X[:, 1], color="blue")
pl.title("Standartized")
pl.show()
```

```python
pl.figure(figsize=(5,5))
pl.scatter(..., X[:, 1], color="blue")
pl.title("Log")
pl.show()
```

```python
pl.figure(figsize=(5,5))
pl.scatter(..., X[:, 1], color="blue")
pl.title("SQRT")
pl.show()
```
